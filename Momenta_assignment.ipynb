```python
import torch
import torchaudio
import os
from torch.utils.data import Dataset
import torch.nn.functional as F
class AudioDataset(Dataset):
    def __init__(self, data_dir, target_length=66230, transform=None):
        self.data_dir = data_dir
        self.target_length = target_length
        self.transform = transform
        self.file_list = [f for f in os.listdir(data_dir) if f.endswith('.flac')]

    def __len__(self):
        return len(self.file_list)

    def __getitem__(self, idx):
        file_path = os.path.join(self.data_dir, self.file_list[idx])
        waveform, sample_rate = torchaudio.load(file_path)

        if waveform.ndimension() == 1:
            waveform = waveform.unsqueeze(0)

        if self.transform:
            waveform = self.transform(waveform)

        if waveform.size(1) < self.target_length:
            pad_size = self.target_length - waveform.size(1)
            waveform = F.pad(waveform, (0, pad_size))
        elif waveform.size(1) > self.target_length:
            waveform = waveform[:, :self.target_length]



        label = 1 if "bonafide" in file_path else 0

        return waveform, torch.tensor(label, dtype=torch.long)
```

```python
from torch.utils.data import DataLoader

def collate_fn(batch):
    waveforms, labels = zip(*batch)
    waveforms = torch.stack(waveforms)
    labels = torch.tensor(labels, dtype=torch.long)
    return waveforms, labels

train_dataset = AudioDataset(data_dir="LA/ASVspoof2019_LA_train/flac",target_length=66230)
print("Number of samples:", len(train_dataset))
print("First sample:", train_dataset[0])
train_loader = DataLoader(train_dataset, batch_size=5, shuffle=True,collate_fn=collate_fn)

val_dataset = AudioDataset("LA/ASVspoof2019_LA_eval")
val_loader = DataLoader(val_dataset, batch_size=5, shuffle=False,collate_fn=collate_fn)
```

```python
import json
import torch
import torch.optim as optim
from torch.utils.data import DataLoader
from aasist.models.AASIST import Model
# Load configuration
with open("config/AASIST-L.conf", "r") as config_file:
    config = json.load(config_file)

# Model Configuration
model_config = config["model_config"]
model = Model(model_config)

# Optimizer Configuration
optim_config = config["optim_config"]
optimizer = optim.Adam(
    model.parameters(),
    lr=optim_config["base_lr"],
    betas=tuple(optim_config["betas"]),
    weight_decay=optim_config["weight_decay"]
)

if config["loss"] == "CCE":
    criterion = torch.nn.CrossEntropyLoss()

# Training Loop
num_epochs = config["num_epochs"]
batch_size = config["batch_size"]
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

for epoch in range(num_epochs):
    model.train()
    optimizer.zero_grad()
    for batch_idx, (data, target) in enumerate(train_loader):
        torch.cuda.empty_cache()
        data, target = data, target
        print("Data shape before model:", data.shape)
        data = data.squeeze(1)
        optimizer.zero_grad()
        output = model(data)
        loss = criterion(output[0], target)

        loss.backward()
        optimizer.step()

        # Optionally, log progress
        if batch_idx % 100 == 0:
            print(f'Epoch {epoch}, Batch {batch_idx}, Loss: {loss.item()}')
```
